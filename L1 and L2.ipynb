```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer()
```

```python
df.iloc[:,[1,11]] = si.fit_transform(df.iloc[:,[1,11]])
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
catcols = df.select_dtypes(object).columns
```

```python
oe = OrdinalEncoder()
```

```python
df[catcols] = oe.fit_transform(df[catcols])
```

```python
df.head()
```

```python
x = df.drop(columns=['price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.3,random_state=1)
```

```python
xtrain
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()

model = lr.fit(xtrain,ytrain)
pred = model.predict(xtest)

train = model.score(xtrain,ytrain)
test = model.score(xtest,ytest)

print(f"Train accuracy {train}\nTest accuracy {test}")
```

```python
model.coef_
```

```python
from sklearn.linear_model import Ridge, Lasso
```

# Ridge

```python
rg = Ridge(alpha=10)
```

```python
model = rg.fit(xtrain,ytrain)
rpred = model.predict(xtest)
```

```python
train = model.score(xtrain,ytrain)
test = model.score(xtest,ytest)

print(f"Train accuracy {train}\nTest accuracy {test}")
```

```python
model.coef_
```

# Lasso

```python
ls = Lasso(alpha=10)

model = ls.fit(xtrain,ytrain)
lspred = model.predict(xtest)
```

```python
train = model.score(xtrain,ytrain)
test = model.score(xtest,ytest)

print(f"Train accuracy {train}\nTest accuracy {test}")
```

# Hyperparameter tuning

```python
for i in range(150,200):
    ls = Lasso(alpha=i)
    ls.fit(xtrain,ytrain)
    train = ls.score(xtrain,ytrain)
    test = ls.score(xtest,ytest)
    
    print(f"{i} {train} {test}")
```

```python
ls = Lasso(alpha=170)

model = ls.fit(xtrain,ytrain)

pred = model.predict(xtest)
```

```python
train = model.score(xtrain,ytrain)
test = model.score(xtest,ytest)

print(f"Train accuracy {train}\nTest accuracy {test}")
```
